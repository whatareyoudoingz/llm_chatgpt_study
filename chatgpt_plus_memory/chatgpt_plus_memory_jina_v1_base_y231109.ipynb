## llm 객체 생성 및 임의의 질의 응답 생성

```python
from langchain.chat_models import ChatOpenAI
from langchain.chains import ConversationChain
from langchain.memory import ConversationSummaryBufferMemory
# 객체 생성
llm = ChatOpenAI(temperature=0,               # 창의성 (0.0 ~ 2.0) #사용할 샘플링 온도 (높을수록 더 무작위하게 출력, 낮을수록 더 집중적이고 결정론적으로 추력)
                 max_tokens=2048,             # 최대 토큰수
                 model_name=MODEL,  # 모델명
                )
memory = ConversationSummaryBufferMemory(llm=llm, max_token_limit=100)

conversation = ConversationChain(
    llm=llm,
    memory = memory
)

# 질의내용
question = '안녕? 나는 아름이라고 해. 나는 2002년 3월 2일에 태어났어. 그리고 겨울을 좋아해.'
conversation.predict(input=question)
```

### chatgpt는 영어로 요약되므로 한글로 요약해서 저장해보자.
- 한글로 요약할 수 없다. 정확히는 찾지 못했다.
- ChatOpenAI, ConversationChain, ConversationSummaryBufferMemory의 함수나 파라미터에 한글로 설정해주는 것이 없었다.
- 다른 메모리를 사용해서 대화를 요약한 후 저장하는 방법이 최선인 것 같다.
- ChatOpenAI : https://api.python.langchain.com/en/latest/chat_models/langchain.chat_models.openai.ChatOpenAI.html
- ConversationChain : https://js.langchain.com/docs/api/chains/classes/ConversationChain
- ConversationSummaryBufferMemory : https://js.langchain.com/docs/api/memory/classes/ConversationSummaryBufferMemory

```python
memory.clear() #moving_summary_buffer부분이 리셋됨.
```

```python
#OpenAI GPT-3를 사용하여 한글로 번역 수행
result=memory.load_memory_variables({})
print(result)
response = openai.ChatCompletion.create(
    model=MODEL, messages=[{"role": "system", "content": "You are a helpful assistant."},
                           {"role": "user", "content": f'Translate the following English text to Korean: "{result}"'},
                           ])
# 번역된 결과 출력
translated_text = response['choices'][0]['message']['content']
print(translated_text)
memory.save_context({"input": ""},
                    {"output": f"{translated_text}"})
```

# ConversationTokenBufferMemory
- 마지막 n토큰 기억
- n을 넘어간다면 마지막 대화를 기준으로 n토큰 기억 (n개가 대화 중간에서 끊긴다면 그 대화는 기억하지 않음.)

```python
from langchain.chat_models import ChatOpenAI
from langchain.chains import ConversationChain
from langchain.memory import ConversationTokenBufferMemory
# 객체 생성
llm = ChatOpenAI(temperature=0,               # 창의성 (0.0 ~ 2.0) #사용할 샘플링 온도 (높을수록 더 무작위하게 출력, 낮을수록 더 집중적이고 결정론적으로 추력)
                 max_tokens=2048,             # 최대 토큰수
                 model_name=MODEL,  # 모델명
                )
memory = ConversationTokenBufferMemory(llm=llm, max_token_limit=200)

conversation = ConversationChain(
    llm=llm,
    memory = memory
)

# 질의내용
question = '안녕? 나는 아름이라고 해. 나는 2002년 3월 2일에 태어났어. 그리고 겨울을 좋아해.'
conversation.predict(input=question)
```

```python
schedule = "내일 아침 8시에는 퓨리랑 미팅이 있습니다. \
어벤져스와 가디언즈 오브 갤럭시 팀의 관계에 대해 상의할 예정입니다. \
12시에는 페퍼와 점심식사가 있습니다. 치즈버거를 먹기로 했습니다. \
3시에는 해피와 스파이더맨 교육 계획 회의가 있습니다 \
5시에는 아크 원자로 개선 버전 공개 행사가 있습니다. \
간단한 소개 자료를 준비해 놓았습니다."
print(len(schedule))
memory.save_context({"input": "내일 일정에 대해 알려줘."},
                    {"output": f"{schedule}"})
conversation.predict(input="자기 전에 확인할 게 있어.")
print(memory.buffer)
print("#############################")
conversation.predict(input="내일 일정에 대해 알려줘.")
print(memory.buffer)
print("#############################")
memory.load_memory_variables({})
```

# ConversationBufferWindowMemory
- 마지막 n개의 대화만 기억

```python
from langchain.chat_models import ChatOpenAI
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferWindowMemory
# 객체 생성
llm = ChatOpenAI(temperature=0,               # 창의성 (0.0 ~ 2.0) #사용할 샘플링 온도 (높을수록 더 무작위하게 출력, 낮을수록 더 집중적이고 결정론적으로 추력)
                 max_tokens=2048,             # 최대 토큰수
                 model_name=MODEL,  # 모델명
                )
memory = ConversationBufferWindowMemory(k=1)

conversation = ConversationChain(
    llm=llm,
    memory = memory
)

# 질의내용
question = '안녕? 나는 아름이라고 해. 나는 2002년 3월 2일에 태어났어. 그리고 겨울을 좋아해.'
conversation.predict(input=question)
```
