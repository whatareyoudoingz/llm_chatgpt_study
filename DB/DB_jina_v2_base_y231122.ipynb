# 데이터베이스 설정
  - sqlite 일때
      - LOCAL_DB_PATH=data/default_sqlite.db
      - LOCAL_DB_TYPE=sqlite # Database Type default:sqlite

  - MYSQL 일때
    - LOCAL_DB_TYPE=mysql
    - LOCAL_DB_USER=root
    - LOCAL_DB_PASSWORD=##mysql root 비밀번호##
    - LOCAL_DB_HOST=127.0.0.1
    - LOCAL_DB_PORT=3306
